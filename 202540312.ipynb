```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()), options = Options())
```

```python
url = "https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EC%B9%98%ED%82%A8"
driver.get(url)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
search = driver.find_element(By.XPATH, '//*[@id="nx_query"]')
# search.click()
# search.clear()
# search.send_keys("피자")
# search.send_keys(Keys.ENTER)
search = driver.find_element(By.TAG_NAME, "body")
for i in range(100):
    search.send_keys(Keys.END)
    time.sleep(1)
```

```python
soup = bs(driver.page_source, "lxml")
```

```python
urls = soup.find_all("a", href = True)
u = []
for i in urls:
    if "news" in str(i):
        u.append(i["href"])
u
```
