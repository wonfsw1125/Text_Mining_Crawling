```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from tqdm import tqdm
import random 

driver = webdriver.Chrome(executable_path="C:/chromedriver.exe")
url = 'https://search.naver.com/search.naver?where=news&query=%EC%B9%98%ED%82%A8&sm=tab_tmr&nso=so:r,p:all,a:all&sort=0'
driver.get(url)
```

```python
# 상호작용
search = driver.find_element(By.XPATH, '//*[@id="nx_query"]') # 검색창 찾기
search.click()

search.clear()
keyword = "피자"
search.send_keys(keyword)
search.send_keys(Keys.ENTER)
```

```python
search = driver.find_element("tag name", "body")
search.send_keys(Keys.PAGE_DOWN)
time.sleep(1)
```

```python
soup = bs(driver.page_source, 'lxml') # 캡쳐
urls = soup.find_all("a", href = True)

result = []
for i in urls:
#     result.append(i["href"])
    try:
        result.append(i["href"])
    except:
        pass
result
```
